```python
import pandas as pd

# Define conversion function, handling single digits, tens, and hundreds, ensuring string type
def convert_follower_count(follower_str):
    follower_str = str(follower_str)  # Ensure string type

    # Extract the numeric part before the conditional check
    number_str = follower_str[:-1] if follower_str[-1] in "mbk" else follower_str
    number = float(number_str)

    # Apply unit conversion
    if follower_str[-1] == "m":
        number *= 1e6
    elif follower_str[-1] == "b":
        number *= 1e9
    elif follower_str[-1] == "k":
        number *= 1e3

    # Round to nearest integer before converting to ensure accuracy:
    number = round(number)

    return int(number)


# Apply the conversion function to the "Followers" column
df["avg_likes"] = df["avg_likes"].apply(convert_follower_count)
```

```python
import pandas as pd
from scipy.stats import pearsonr

# Assuming your DataFrame is named 'df'

# Define a function to interpret the strength of correlation
def interpret_correlation_strength(corr_value):
    if abs(corr_value) >= 0.7:
        return "strong"
    elif abs(corr_value) >= 0.3:
        return "moderate"
    else:
        return "weak"

# Compute Pearson correlation coefficient and p-value for Followers and Total Likes
corr_followers_total_likes, p_value_followers_total_likes = pearsonr(df['Followers'], df['total_likes'])
print(f"Correlation between Followers and Total Likes: {corr_followers_total_likes:.2f}")
print(f"P-value: {p_value_followers_total_likes:.4f}")
if p_value_followers_total_likes < 0.05:
    print("The correlation is statistically significant.")
    print(f"The correlation is considered {interpret_correlation_strength(corr_followers_total_likes)}.")
else:
    print("The correlation is not statistically significant.")
print("\n")

# Compute Pearson correlation coefficient and p-value for Followers and Influence Score
corr_followers_influence_score, p_value_followers_influence_score = pearsonr(df['Followers'], df['influence_score'])
print(f"Correlation between Followers and Influence Score: {corr_followers_influence_score:.2f}")
print(f"P-value: {p_value_followers_influence_score:.4f}")
if p_value_followers_influence_score < 0.05:
    print("The correlation is statistically significant.")
    print(f"The correlation is considered {interpret_correlation_strength(corr_followers_influence_score)}.")
else:
    print("The correlation is not statistically significant.")
print("\n")

# Compute Pearson correlation coefficient and p-value for Posts and Average Likes
corr_posts_avg_likes, p_value_posts_avg_likes = pearsonr(df['Posts'], df['avg_likes'])
print(f"Correlation between Posts and Average Likes: {corr_posts_avg_likes:.2f}")
print(f"P-value: {p_value_posts_avg_likes:.4f}")
if p_value_posts_avg_likes < 0.05:
    print("The correlation is statistically significant.")
    print(f"The correlation is considered {interpret_correlation_strength(corr_posts_avg_likes)}.")
else:
    print("The correlation is not statistically significant.")
print("\n")

# Compute Pearson correlation coefficient and p-value for Posts and Influence Score
corr_posts_influence_score, p_value_posts_influence_score = pearsonr(df['Posts'], df['influence_score'])
print(f"Correlation between Posts and Influence Score: {corr_posts_influence_score:.2f}")
print(f"P-value: {p_value_posts_influence_score:.4f}")
if p_value_posts_influence_score < 0.05:
    print("The correlation is statistically significant.")
    print(f"The correlation is considered {interpret_correlation_strength(corr_posts_influence_score)}.")
else:
    print("The correlation is not statistically significant.")
```

```python
import pandas as pd
from scipy.stats import poisson
from scipy.stats import chi2_contingency

# Assuming your DataFrame is named 'df' and has a column 'Posts'
# Replace 'Posts' with the actual column name in your DataFrame

# Generate a sample from a Poisson distribution (for demonstration purposes)
# You would replace this line with your actual data
sample_data = poisson.rvs(mu=df['Posts'].mean(), size=len(df['Posts']))

# Create a contingency table for the observed and expected frequencies
observed_freq = pd.value_counts(sample_data, sort=False)
expected_freq = poisson.pmf(observed_freq.index, mu=df['Posts'].mean()) * len(df['Posts'])

# Perform the Chi-square test for goodness of fit
chi2_stat, p_value, dof, expected = chi2_contingency([observed_freq, expected_freq])

# Set significance level
alpha = 0.05

# Check if the null hypothesis is rejected
if p_value < alpha:
    print("The data does not follow a Poisson distribution.")
else:
    print("The data follows a Poisson distribution.")

# You can also print the Chi-square statistic, p-value, and degrees of freedom for further analysis
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
```

```python
import pandas as pd
from scipy.stats import anderson

# Assuming your DataFrame is named 'df' and has a column 'Posts'
# Replace 'Posts' with the actual column name in your DataFrame

# Perform the Anderson-Darling test for exponential distribution
result = anderson(df['Posts'], dist='expon')

# Extract the test statistic and p-value
statistic = result.statistic
p_value = result.significance_level[-1]  # P-value associated with the largest critical value

# Print the test result
print(f"Anderson-Darling Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Check if the null hypothesis is rejected
alpha = 0.05
if p_value < alpha:
    print("The data does not follow an exponential distribution.")
else:
    print("At the 5% significance level, the data follows an exponential distribution.")
```

```python
import pandas as pd
from scipy.stats import poisson
from scipy.stats import chi2_contingency

# Assuming your DataFrame is named 'df' and has a column 'Posts'
# Replace 'Posts' with the actual column name in your DataFrame

# Calculate observed frequencies
observed_freq = pd.value_counts(df['Posts'], sort=False)

# Calculate expected frequencies based on Poisson distribution
expected_freq = poisson.pmf(observed_freq.index, mu=df['Posts'].mean()) * len(df['Posts'])

# Perform the Chi-square test for goodness of fit
chi2_stat, p_value, dof, expected = chi2_contingency([observed_freq, expected_freq])

# Set significance level
alpha = 0.05

# Print the test results
print(f"Chi-square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")

# Check if the null hypothesis is rejected
if p_value < alpha:
    print("The data does not follow a Poisson distribution.")
else:
    print("The data follows a Poisson distribution.")
```

```python
import pandas as pd
from scipy.stats import shapiro

# Assuming your DataFrame is named 'df' and has a column 'Posts'
# Replace 'Posts' with the actual column name in your DataFrame

# Perform the Shapiro-Wilk test for normality
statistic, p_value = shapiro(df['Posts'])

# Set significance level
alpha = 0.05

# Print the test results
print(f"Shapiro-Wilk Test Statistic: {statistic}")
print(f"P-value: {p_value}")

# Check if the null hypothesis is rejected
if p_value < alpha:
    print("The data does not follow a normal distribution.")
else:
    print("The data follows a normal distribution.")
```

```python
import pandas as pd
from scipy.stats import skew

# Assuming your DataFrame is named 'df' and has a column 'Posts'
# Replace 'Posts' with the actual column name in your DataFrame

# Calculate the skewness
skewness = skew(df['Posts'])

# Print the skewness value
print(f"Skewness for 'Posts' column: {skewness}")

# Interpret the skewness value
if skewness > 0:
    print("The distribution is positively skewed.")
elif skewness < 0:
    print("The distribution is negatively skewed.")
else:
    print("The distribution is approximately symmetric.")
```
